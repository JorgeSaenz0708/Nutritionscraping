```python
# Import necessary libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
# Set the main URL where are located the categories 
url = "https://www.fatsecret.cl/calor%C3%ADas-nutrici%C3%B3n/grupo/bebidas"
data = requests.get(url).text
```

```python
# Parse htlm data 
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
# Create a list from all categories of main web page - by class 
alimentos_list = soup.find_all( class_ = "borderBottom" )

alimentos_list
```

```python
# Gather all url and fill a list with them from each category Saco 
alimentos_list_links = []
for link_alimentos in alimentos_list:
    l_a = link_alimentos.find('a')
    alimentos_list_links.append(l_a['href'])
alimentos_list_links
```

```python
# Create a data frame to fill then
noms_cols = ["Nombre", "Energía", "Proteína", "Grasa", "Grasa Saturada", "Grasa Poliinsaturada", "Grasa Monoinsaturada",
             "Grasa Trans","Colesterol", "Carbohidratos", "Fibra", "Azúcar", "Sodio", "Potasio" ]

df = pd.DataFrame(columns = noms_cols)

# function to convert list into dictionary that will allow us to set data into the data frame according to columns name
def Convert(a):
    it = iter(a)
    res_dct = dict(zip(it, it))
    return res_dct
```

```python
# Gather Full data from sub-categories nesting two for loops

# Go into each link to find link to sub categories and then create a new url address 

for L in range(0,len(alimentos_list_links)):

    n_url = "https://www.fatsecret.cl" + alimentos_list_links[L] # index will change trought the links
    n_data = requests.get(n_url).text
    n_soup = BeautifulSoup(n_data, 'html.parser')

    # got all links in the page
    links = n_soup.find_all( class_ = "food_links" )

    #total of Sub links of each sub categories
    list_links = []
    for c in range(0, len(links)):
        for link in links[c].find_all('a'):
            h = link.get('href')
            list_links.append(h)
    
    


    ##################################          Here start the data of each sub category 

    for sub_k in range(0,len(list_links)):

        # Scrape fodd page by each sub category
        sub_url = "https://www.fatsecret.cl" + list_links[sub_k] # sub category
        time.sleep(2) # time sleep put because Data took time to be gathered so to avoid errors. It can be enhaced using try exceptions.
        sub_data = requests.get(sub_url).text
        sub_soup = BeautifulSoup(sub_data, 'html.parser')

        # Extract food name by tag
        nom_alimento = sub_soup.find('h1').text
        nom_alimento = str(nom_alimento)

        # Find tables by tag and class
        table = sub_soup.find('table', class_='generic')

        # On each table gather the nutritional data by class name
        macros = table.find_all( class_ = "factPanel" )

        # Organize the wished info 
        macros = macros[0].text.strip().split('\n')

        # Filter only data we want - using index position
        macros = macros[0:(macros.index("Origen: FatSecret Platform API")-1)]

        # Delete all blank spaces
        while macros.count("") >= 1:
            macros.remove("")


        # Gather energy info 
        calorias = macros[6]

        # Using the function created before to convert each list into dictionary to then fill data frame by columns name
        # and we update two more columns in order to get full data frame
        dictionary = Convert(macros[7:])
        dictionary.update({"Nombre" : nom_alimento})
        dictionary.update({"Energía" : calorias})

       # Fill data frame from dictionary 
        df = df.append(dictionary,  ignore_index=True)



df
```
